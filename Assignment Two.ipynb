### 1. Using Pandas, read the cars.csv 

```python
import numpy as np
import pandas as pd
cars = pd.read_csv("cars.csv")
cars
```

```python
cars.plot.scatter(x = "hp", y = "mpg")
```

The hp and mpg are correlated, however, they are negatively correlated. As hp increases, mpg decreases. 

### 4. Perform the simple linear regression on the variables hp and mpg

```python
from sklearn import linear_model
```

```python
X = np.array(cars['hp']).reshape(-1, 1)
Y = np.array(cars['mpg']).reshape(-1, 1)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, Y)
```

### 5. Display the results for the intercept, coefficients, and R-Squared

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X,Y)
```

Y = 30.09886054 - .06822828x + e

### 6. Predict the mpg for a car with an hp of 100

```python
reg.predict([[100]])
reg.predict(X)
```

### 7. Calculate the Mean Absolute Error (MAE) and interpret the results

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y, reg.predict(X))
```

The MAE is the average error that we get when we do a prediction. Also, the MAE is absolute so it can be either negative or positive. Based on the MAE we calculated here, our error is fairly large as the difference between of 17 and 20 is rather significant.
